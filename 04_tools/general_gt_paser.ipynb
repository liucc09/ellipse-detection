```python
raw_path = r'/home/liucc/Workspace/02_Ellipse/03_results/05_general/04_ellipse_label_temp'
out_path = r'/home/liucc/Workspace/02_Ellipse/03_results/05_general/04_ellipse_label'
```

```python
import os 
fnames = os.listdir(raw_path)
fnames = [n[:-4] for n in fnames if n[-3:]=='txt']
fnames.sort()
print(fnames[:10])
```

```python
print(fnames[0][3:-4])
```

```python
import numpy as np
for name in fnames:
    full_in = raw_path+'/'+name+'.txt'
    full_out = out_path+'/'+name[3:-4]+'.txt'
    with open(full_in,'r') as f:
        content = f.read()
        lines = content.split('\n')
        params = [[float(v) for v in l.split('\t')] for l in lines if len(l.split('\t'))==5]
        params = np.array(params)
        #params[:,-1] = params[:,-1]*180/np.pi
        
    with open(full_out,'w') as f:
        estr = '\n'.join([' '.join(['%.4f'%v for v in l]) for l in params])
        f.write(estr)
        
```
