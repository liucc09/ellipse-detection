```python
from cc_ellipse import *
import os
```

```python
dataset = '03_brightness'
raw_path  = f'/home/liucc/Workspace/02_Ellipse/03_results/{dataset}/01_raw/'
base_path = f'/home/liucc/Workspace/02_Ellipse/03_results/{dataset}/02_transfer/'
```

```python
import os 
fnames = os.listdir(base_path)
fnames = [n[:-4] for n in fnames if n[-3:]=='png']
fnames.sort()
print(fnames)
```

```python
import sys
sys.path.append('/home/liucc/Workspace/02_Ellipse/01_data')
from cc_process_image import *
import matplotlib.pyplot as plt
import imageio
import time

#全局参数
GLOB = {
    'MIN_LEN' : 70,
    'MIN_COVER_ANGLE' : 110,
    'MIN_COVER_RHO1' : 0.5,
    'MIN_COVER_RHO2' : 0.2,
}

total_time = 0

prs = []
ims = []
raws = []

for name in fnames:
    full_path = base_path+name+'.png'
    fraw_path = raw_path+name+'.png'
    #oname = result_path+name+'.txt'
    
    start = time.time()
    
    im = imageio.imread(full_path)
    raw = imageio.imread(fraw_path)
    
    im = postprocess_image(im,0.3)
    ells = detect_ellipses_im(im,GLOB)
    
    end = time.time()
    total_time+=end-start
    
    ims.append(im)
    raws.append(raw)
    prs.append(ells)
        
    
print('Average Time:%.4f' % (total_time/len(fnames)))
```

```python
num = 40
_,axes = plt.subplots(nrows=14,ncols=3,figsize=(20,60)) 

axes = axes.reshape(-1)

for ax,ells,im in zip(axes,prs[:num],raws[:num]):
    for ell in ells:
        exy = draw_ellipse(*ell)
        ax.plot(exy[0,:],exy[1,:],linewidth='1',color='red')
        
    ax.imshow(im,cmap='gray',vmin=0,vmax=255)

plt.show()
```
