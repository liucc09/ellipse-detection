# 生成目标图

```python
%matplotlib inline
import matplotlib.pyplot as plt
from skimage.filters import gaussian
from skimage.color import rgb2hsv
im0 = imxs[0]
im1 = gaussian(im0,sigma=1)

_,axes=plt.subplots(ncols=2,figsize=(20,10))
axes[0].imshow(im0)
axes[1].imshow(im1)

plt.show()
```

```python
from tqdm import tqdm_notebook as tqdm
from skimage.transform import rotate
from skimage.color import rgb2hsv

width = 256
gap = 128

im_xs = []
im_ys = []

itype = 'gray'

for i,(imx,imy) in tqdm(enumerate(zip(imxs,imys))):

    if itype=='gray':
        h,w = imx.shape
    else:
        h,w,_ = imx.shape
    
    h1,w1 = imy.shape
    
    assert h==h1 and w==w1, "图片尺寸不匹配"

    for r in range(0,max(h-width,1),gap):
        for c in range(0,max(w-width,1),gap):
            
            
            r = max(min([r,h-width]),0)
            c = max(min([c,w-width]),0)
                
            if itype=='gray':
                xt = imx[r:r+width,c:c+width].astype(dtype=np.float32)
            else:
                xt = imx[r:r+width,c:c+width,:].astype(dtype=np.float32)
            
            yt = imy[r:r+width,c:c+width].astype(dtype=np.float32)
            
            if itype=='gray':
                x = np.zeros((width,width),dtype=np.float32)
            else:
                x = np.zeros((width,width,3),dtype=np.float32)
            
            y = np.zeros((width,width),dtype=np.float32)
            
            if itype=='gray':
                hx,wx = xt.shape
            else:
                hx,wx,_ = xt.shape
            
            if itype=='gray':
                x[:hx,:wx] = xt
            else:
                x[:hx,:wx,:] = xt
            
            y[:hx,:wx] = yt
            
            if itype=='gray':
                assert x.shape[0]==256 and x.shape[1]==256
            else:
                assert x.shape[0]==256 and x.shape[1]==256 and x.shape[2]==3

            thr = np.mean(y)+10
            y = np.where(y>thr,1,0)

            if np.sum(y)>1000:
                im_xs.append(x)
                im_ys.append(y)
                    
```
